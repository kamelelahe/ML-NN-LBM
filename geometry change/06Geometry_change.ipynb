
#Functions


```python
def Change_Singel(MAT,resol):
  total=resol*resol
  Avg=np.mean(MAT[MAT>0])
  AFT=np.ones([resol,resol])
  i=0
  j=0
  grain=0
  while i<resol:
    j=0
    while j<resol:
      a=MAT[i,j]
      if a<0.6*Avg:
          AFT[i][j]=0
      j=j+1  
    i=i+1
  return AFT
```

```python
def Change_Singel_Omega(MAT,resol,Omega):
  total=resol*resol
  AFT=np.ones([resol,resol])
  i=0
  j=0
  grain=0
  while i<resol:
    j=0
    while j<resol:
      a=MAT[i,j]
      if a<Omega:
          AFT[i][j]=0
      j=j+1  
    i=i+1
  return AFT
```

```python
def Porosity_Single(MAT,resol):
  total=resol*resol
  i=0
  j=0
  grain=0
  while i<resol:
    j=0
    while j<resol:
      a=MAT[i,j]
      if a==0:
          grain=grain+1
      j=j+1  
    i=i+1
  poro=1-(grain/total)
  return poro
```

```python
def PorosityCalc(MAT,numfiles,resol):
  total=resol*resol
  k=0
  poro=[]
  for k in np.arange(0,numfiles-1):
    i=0
    j=0
    grain=0
    print(k)
    while i<resol:
      j=0
      while j<resol:
        a=MAT[i,j,k]
        if a==0:
            grain=grain+1
        j=j+1  
      i=i+1

    poro.append(1-(grain/total))
  return poro
```

```python
def Segment(MAT,numfiles,resol):
  total=resol*resol
  k=0
  AFT=np.zeros([resol,resol,numfiles])
  for k in np.arange(0,numfiles-1):
    i=0
    j=0
    grain=0
    print(k)
    while i<resol:
      j=0
      while j<resol:
        a=MAT[i,j,k]
        if a==0:
            AFT[i][j][k]=1
        j=j+1  
      i=i+1
  return AFT
```

```python
def imageWriter(MAT,name):
  cmap = plt.cm.jet
  image = cmap(MAT)
  name1=name+"0000.bmp"
  name2=name+"0001.bmp"
  name3=name+"0002.bmp"
  plt.imsave(name1, image,cmap='Greys',vmin=0, vmax=255)
  plt.imsave(name2, image,cmap='Greys',vmin=0, vmax=255)
  plt.imsave(name3, image,cmap='Greys',vmin=0, vmax=255)
```
